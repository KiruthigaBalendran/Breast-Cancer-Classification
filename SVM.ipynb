```python
import pandas as pd
import numpy as np

# data visualization
import matplotlib.pyplot as plt
import seaborn as sns

# machine learning
from sklearn.preprocessing import StandardScaler

import sklearn.linear_model as skl_lm
from sklearn.preprocessing import LabelEncoder
from sklearn import neighbors
from sklearn.metrics import confusion_matrix, classification_report, precision_score
from sklearn.model_selection import train_test_split
from sklearn import svm


# package settings initialization
sns.set(style="whitegrid", color_codes=True, font_scale=1.3)

%matplotlib inline
```

# Read the Data

```python
# read the data and check the first 5 rows
df = pd.read_csv('/content/drive/MyDrive/MSc/Semester-4/Pattern-Recognition/Assignments/SVM/data.csv', index_col=0)
df.head()
```

# Data Summary

```python
# general summary of the dataframe
df.info()
```

# Pre-processing

```python
# There aren't any missing values except the column Unnamed: 32. So this column can be removed
df = df.drop('Unnamed: 32', axis=1)
```

```python
# Using Labelencoder to transform diagnosis characters to digits
labelencoder = LabelEncoder()
df['diagnosis'] = labelencoder.fit_transform(df['diagnosis'])
X = df
y = df['diagnosis']


# Split the data into training and testing sets
# The paper has split the data as 65.10% as training and 34.90% as testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.349, random_state=40)
```

# SVM Model

```python
# Create a svm Classifier using Linear Kernel
svm_model = svm.SVC(kernel='linear')
svm_model.fit(X_train, y_train)
```

# SVM Model Prediction

```python
svm_predictions = svm_model.predict(X_test)
```

# SVM Accuracy 

```python
# Use score method to get accuracy of model
svm_score = svm_model.score(X_test, y_test)
print("Accuracy SVM classifier : " + str(svm_score))
```

# SVM Classification Report and Confusion Matrix

```python
print(classification_report(y_test, svm_predictions, digits=3))
svm_cm = confusion_matrix(y_test, svm_predictions)
print(svm_cm)
```
